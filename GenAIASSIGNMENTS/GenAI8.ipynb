```python
from TTS.api import TTS
tts = TTS("tts_models/multilingual/multi-dataset/xtts_v2", gpu=True)

# generate speech by cloning a voice using default settings
tts.tts_to_file(text="hello my name is Anthony parker , I am the worlds most famous gunda and local gang leader",
                file_path="output.wav",
                speaker_wav="harvard.wav",
                language="en")
```

```python
# Import the necessary module to use ipd
from IPython.display import Audio

# Play the generated audio using Audio
Audio("output.wav")
```
