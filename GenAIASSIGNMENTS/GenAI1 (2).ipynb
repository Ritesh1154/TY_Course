Implement a basic probabilistic generative model (e.g., Gaussian Mixture Models).


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
```

```python
# Generate synthetic data
np.random.seed(42)
X1 = np.random.normal(loc=0, scale=1, size=(300, 2))  # Cluster 1
X2 = np.random.normal(loc=5, scale=1.5, size=(300, 2))  # Cluster 2
X3 = np.random.normal(loc=-3, scale=0.5, size=(300, 2))  # Cluster 3
X = np.vstack([X1, X2, X3])  # Combine into one dataset
```

```python
# Fit Gaussian Mixture Model
num_components = 3
GMM = GaussianMixture(n_components=num_components, covariance_type='full', random_state=42)
GMM.fit(X)
```

```python
# Predict cluster assignments
labels = GMM.predict(X)
```

```python
# Plot the results
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', alpha=0.6)
plt.scatter(GMM.means_[:, 0], GMM.means_[:, 1], c='red', marker='x', s=100, label='Cluster Centers')
plt.title("Gaussian Mixture Model Clustering")`
plt.legend()
plt.show()
```
